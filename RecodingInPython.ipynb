#Recode Activity into a new variable

```python
eating_habits.Activity.value_counts()
```

```python
def activity (series):
    if series == "Eating fruit" :
        return 0
    if series == "Eating candy, chocolate bars" :
        return 0
    if series == "Eating french fries" :
        return 0
    if series == "Eating hamburgers, hot dogs or sausages" :
        return 0
    if series == "Eating whole wheat or rye bread" :
        return 0
    if series == "Drinking soft drinks, cola or other drinks with sugar" :
        return 0
    if series == "Drinking coffee" :
        return 0
    if series == "Eating raw vegetables" :
        return 0
    if series == "Eating potato chips, crisps" :
        return 0
    if series == "Drinking low fat milk" :
        return 0
    if series == "Drinking whole milk" :
        return 0
    if series == "Eating peanuts" :
        return 0
    
    
eating_habits['ActivityR'] = eating_habits['Activity'].apply(activity)

```

#Recode Frequency from text to numbers in the same variable

```python
eating_habits.Frequency.value_counts()
```

```python
cleanup = {"Frequency":{"More than once a day":0, "Once a day":1, "At least once a week":2, "Seldom":3, "Never":4}}
eating_habits.replace(cleanup, inplace=True)
```
